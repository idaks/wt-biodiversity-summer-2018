```python
#replace underscores with blanks
df["t1"] = df["t1"].str.replace('_',' ')
```

```python
df
```

```python
#only show the rows of articulations 
df[df['t1'].str.contains("equals|is included in|includes|overlaps|disjoints")]
```

```python
#create a new dataframe based on the column t1
df1 = df[df['t1'].str.contains("equals|is included in|includes|overlaps|disjoints")]

#split the column based on the rcc-5 relations 
df1 = df1['t1'].str.split('equals|is included in|includes|overlaps|disjoints', 1, expand=True)
```

```python
#update the column names to t1 and t2
df1.columns=["t1","t2"]
```

## 3. Merge the dataset

```python
df1.head()
```

```python
occurrences.head()
```

```python
df1.dtypes
```

```python
occurrences.dtypes
```

```python
#trim leading and trailing whitespaces in t1 column in df1 
df1['t1']=df1['t1'].str.strip()

#make sure the columns we are matching are all strings: 't1' in Euler/X input file vs. 'scientificName' in the occurrence dataset
df1.t1=df1.t1.astype(str)
occurrences.scientificName=occurrences.scientificName.astype(str)
```

```python
#merge the two datasets based on 't1' in Euler/X input file vs. 'scientificName', using LEFT join (occurrence data) 
result=pd.merge(occurrences,df1, left_on='scientificName',right_on='t1',how='left')
```

```python
result
```

```python
#add a new column "newName", and match the column where scientific name = t1, return t2 in the newName column
result['newName']=np.where(result['scientificName']==result['t1'],result['t2'],np.nan)
```

```python
result
```

```python
#drop the unnecessary columns
result.drop(['scientificName','t1','t2'],inplace=True,axis=1)
```

```python
result
```

```python
#update the column name to scientificName
result=result.rename(columns={'newName':'scientificName'})
```

```python
result
```

```python
#drop any rows that contain missing values  
result.dropna(axis=0,how='any',inplace=True)
```

```python
result
```

```python
#save the new result as a new CSV file
result.to_csv('MagnoliaSecChapman1883test.csv')
```
