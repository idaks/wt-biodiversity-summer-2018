```python
#import all the necessary libraries we will be using 
import pandas as pd
import numpy as np
```

## LeanEuler output relations table

```python
#reads into LeanEuler output file
df=pd.read_csv("inputFiles/Georel_3.csv")
```

```python
df
```

```python
#remove punctuation marks 
df['t1'] = df['x1'].str.replace('[^\w\s]','')
df['t2'] = df['x2'].str.replace('[^\w\s]','')
df['relations'] = df['x3'].str.replace('"','')
```

```python
df
```

```python
#remove taxonomy labels
df["t1"] = df['t1'].str.replace('^[^_]+_','')
df["t2"] = df['t2'].str.replace('^[^_]+_','')
```

```python
df
```

```python
#drop unnecessary columns
df.drop(['Unnamed: 0', 'pw', 'x1', 'x2', 'x3'],1, inplace=True)
```

```python
df.head()
```

```python
#remove parent-level nodes
df2=df[df['t1'].str.contains("USA")==False]
df2=df2[df2['t2'].str.contains("USA")==False]
```

```python
df2
```

```python
#if the relations between the two taxonomies are 'equals' or 'is_included_in', made t2 into their new name
a=df2['relations']=="="
b=df2['relations']=="<"
df2['new']=np.where(a|b,df2['t2'],'NA')
```

```python
df2
```

```python
#remove all the 'disjoint' relations, we won't map them in our datasets 
df2=df2[df2['relations']!="!"]
```

```python
df2
```

```python
#Groupby Column t1, if there are duplicates in t2, concatenate them using 'OR'
df2['new'] = df2.groupby(['t1'])['t2'].transform(lambda x: ' OR '.join(x))
```

```python
df2
```

```python
#drop the duplicated rows
df3=df2.drop_duplicates('t1')
```

```python
df3
```

## Merge with dataset

```python
#an actual csv file based on CEN regions 
geo=pd.read_csv("inputFiles/GeoTaxonomyCEN.csv")
```

```python
geo
```

```python
df3.head()
```

```python
df3.dtypes
```

```python
geo.dtypes
```

```python
#trim leading and trailing whitespaces in t1 column in df1 
df3['t1']=df3['t1'].str.strip()

#make sure the columns we are matching are all strings: 't1' in Euler/X input file vs. 'scientificName' in the occurrence dataset
df3.t1=df3.t1.astype(str)
geo.Regions_CEN=geo.Regions_CEN.astype(str)
```

## Merge and Map the dataset to create a new dataset using NDC taxonomy

```python
result=pd.merge(geo,df3, left_on='Regions_CEN',right_on='t1',how='left')
```

```python
result
```

```python
#add a new column "newName", and match the column where scientific name = t1, return t2 in the newName column
result['newName']=np.where(result['Regions_CEN']==result['t1'],result['new'],np.nan)
```

```python
result
```

```python
#drop the unnecessary columns
result.drop(['Regions_CEN','t1','t2','relations','new'],inplace=True,axis=1)
```

```python
#update the column name to Regions_NDC
result=result.rename(columns={'newName':'Regions_NDC'})
```

```python
result
```

```python
#save the new result as a new CSV file
result.to_csv('CEN_NDC_Correct.csv')
```
