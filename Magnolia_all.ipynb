# Create a function to reproduce the code for visualizing co-occurence species distribution

```python
#d= datasets; t= taxonomy name in text format
def plotdata(d,t):
    sns.lmplot( x="decimalLongitude", y="decimalLatitude", data=d, fit_reg=False, hue='scientificName', legend=False, size=10, aspect=1.5)
    plt.legend(loc='lower right') 
    plt.title(t)
    m=Basemap(llcrnrlat = 24.09,
            llcrnrlon = -96.96,
            urcrnrlat = 47.27,
            urcrnrlon = -64.61) #this is where the map's geo-locations are 
    m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
    m.fillcontinents(color='grey', alpha=0.3)
    m.drawcoastlines(linewidth=0.1, color="white")
    plt.savefig(t, bbox_inches='tight') #this line directly saves the figures in PNG format 
```

### Reads csv datsets. Datasets metadata fields in Darwin-core standard. 
Here we only tested 2 datasets. We can read more datasets if desired. 

```python
#d
d1=pd.read_csv("Magnolia/Dataset/geocoded/Magnolia_secWeakley2014_geocoded.csv",error_bad_lines=False)
d2=pd.read_csv("Magnolia/Dataset/geocoded/Magnolia_secChapman1883_geocoded.csv",error_bad_lines=False)
```

### This plot the Magnolia species co-occurrence distributions based on sec.Weakley 2014 and sec.Chapman 1883 taxonomies

```python
plotdata(d1,'Weakley 2014')
plotdata(d2,'Chapman 1883')
```
