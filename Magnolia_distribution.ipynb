```python
#import the libaries we need to use

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import us
%matplotlib inline

##  Import display 
from IPython.display import display


### ipywidget libraries
from ipywidgets import HBox, VBox, IntSlider, Play, jslink
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

### bqplot libraries
import bqplot
from bqplot import (
    Axis, ColorAxis, LinearScale, DateScale, DateColorScale, OrdinalScale, Mercator, Orthographic,
    OrdinalColorScale, ColorScale, Scatter, Lines, Figure, Tooltip, Map, ColorAxis, AlbersUSA, topo_load
)

### seaborn
import seaborn as sns

### basemap
from mpl_toolkits.basemap import Basemap
```

```python
#reads into the location.txt file, provides headers
d1=pd.read_csv("Datasets/Magnolia_secChapman1883_geocodedtest.csv")
d2=pd.read_csv("Datasets/Magnolia_secWeakley2014_geocodedtest.csv")
```

```python
#SecChapman1883 seaborn scatter plot
sns.lmplot( x="decimalLongitude", y="decimalLatitude", data=d1, fit_reg=False, hue='scientificName', legend=False, size=10, aspect=1.5)
plt.legend(loc='lower right')
plt.savefig('Magnolia_scatter1883.png', bbox_inches='tight')


#SecWeakley1994 seaborn scatter plot
sns.lmplot( x="decimalLongitude", y="decimalLatitude", data=d2, fit_reg=False, hue='scientificName', legend=False, size=10, aspect=1.5)
plt.legend(loc='lower right')

plt.savefig('Magnolia_scatter1994.png', bbox_inches='tight')
```

```python
#secChapman1883 basemap
fig = plt.figure(figsize=(25, 25))

m1=Basemap(llcrnrlat = 23.05,
            llcrnrlon = -126.58,
            urcrnrlat = 50.43,
            urcrnrlon = -64.09)
m1.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m1.fillcontinents(color='grey', alpha=0.3)
m1.drawcoastlines(linewidth=0.1, color="white")
x=d1["decimalLongitude"] 
y=d1["decimalLatitude"]
d1['labels_enc'] = pd.factorize(d1['scientificName'])[0]
m1.scatter(x,y,c=d1['labels_enc'],cmap='Set1',marker="o")
labels = pd.unique(d1['scientificName'])

plt.legend(labels,frameon=True, fontsize=12,loc = "lower right", borderpad = 1.8,
title='Chapman1883')

plt.savefig('Magnolia_baseMapall1883.png', bbox_inches='tight')


#secWeakley1994 basemap

fig = plt.figure(figsize=(25, 25))

m2=Basemap(llcrnrlat = 23.05,
            llcrnrlon = -126.58,
            urcrnrlat = 50.43,
            urcrnrlon = -64.09)
m2.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m2.fillcontinents(color='grey', alpha=0.3)
m2.drawcoastlines(linewidth=0.1, color="white")
x=d2["decimalLongitude"] 
y=d2["decimalLatitude"]
d2['labels_enc'] = pd.factorize(d2['scientificName'])[0]
m2.scatter(x,y,c=d2['labels_enc'],cmap='Set1',marker="o")
labels = pd.unique(d2['scientificName'])

plt.legend(labels,frameon=True, fontsize=12,loc = "lower right", borderpad = 1.8,
title='Weakley2014')

plt.savefig('Magnolia_baseMapall1994.png', bbox_inches='tight')

```
